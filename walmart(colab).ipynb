```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '', regex=False).astype(float)
df.dtypes
```

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
df['total price'] = df['unit_price'] * df['quantity']
display(df.head())
```
